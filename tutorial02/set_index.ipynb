```python
import yfinance as yf
import time
# Define the ticker symbol for Thailand Set Index
# Get the data of the stock
ticker = yf.Ticker('^SET.BK')

# Fetch and update stock values
df = ticker.history(period='8d', interval='1m').reset_index()
df.info()
df.head()
```

```python
# df['retrieval_date']=df['Datetime'].dt.date
from datetime import datetime
df['retrieval_date']= datetime.now().date()
df.info()
df.head()
```

```python
# Write DataFrame to a directory "output_parquet" partitioned by retrieval_time
df.to_parquet(
    "output_setindex",
    partition_cols=["retrieval_date"],   # <-- crucial for partitioning by retrieval_time
    engine="pyarrow",
    index=False
)
```
